## Домашнее задание
Написать приложение, которое собирает основные новости с сайтов mail.ru, lenta.ru.
Для парсинга использовать xpath. Структура данных должна содержать:
* название источника,
* наименование новости,
* ссылку на новость,
* дата публикации

```python
import requests
import pandas as pd
from lxml import html
from pprint import pprint
```

```python
header = {'user-agent':'Mozilla/5.0 (Windows NT 6.1; Win64; x64) '
          'AppleWebKit/537.36 (KHTML, like Gecko) '
          'Chrome/78.0.3865.120 Safari/537.36'}

URL_mail = 'https://mail.ru/?from=m'
URL_lenta = 'https://m.lenta.ru'
```

```python
url_list = []
txt_list = []
url_main = []

req = requests.get(URL_mail, headers=header)
root = html.fromstring(req.text)
url = root.xpath("//a[@class='list__item']/@href")
txt = root.xpath("//a[@class='list__item']/div[@class='list__item__desc']/span[@class='list__item__title']/text()")

for i in range(len(url)):
    url_list.append(url[i])
    txt_list.append(txt[i])
    url_main.append('MAIL.RU')

req = requests.get(URL_lenta, headers=header)
root = html.fromstring(req.text)
url = root.xpath("//li[@class='b-list-item b-list-item_news']/a[@class='b-list-item__link']/@href")
txt = root.xpath("//span[@class='b-list-item__title']/text()")

for i in range(len(url)):
    url_list.append('m.lenta.ru' + url[i])
    txt_list.append(txt[i])
    url_main.append('LENTA.RU')
```

```python
dictionary = {'Text_news':txt_list, 'URL_news':url_list, 'URL_main':url_main}
df = pd.DataFrame(data=dictionary)
```
